## Visualizing what answers my model tends to pick

```python
import pickle

name = 'answers_model_likes'
with open('../Data_and_Models/' + name + '.pkl', 'rb') as f:
    answers_model_likes = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_context("poster")
sns.set_style("dark")

plt.hist(answers_model_likes['where_good'],bins=10,range=(0,10), normed=True)
plt.xticks(np.arange(1,11)-0.5,np.arange(1,11))
plt.xlabel('Answer Number')

plt.tight_layout()
plt.show()
```

## What about when the model doesn't pick the first answer

```python
x = np.array(answers_model_likes['where_good'])[np.where(answers_model_likes['skip_bool'])[0]]

plt.hist(x,bins=10,range=(0,10), normed=True)
plt.xticks(np.arange(1,11)-0.5,np.arange(1,11))
plt.xlabel('Answer Number')

plt.tight_layout()
plt.show()
```
